```python
filtered_df_zip = filtered_df[filtered_df['9. ZIP'].astype(str).str.len() > 5].copy()
filtered_df_zip

# filtered_df_zip['9. ZIP'] = filtered_df_zip['9. ZIP'].astype(int)
```

```python
# Convert to string and store full ZIP for splitting
filtered_df['ZIP_str'] = filtered_df['9. ZIP'].astype(str)

# First 5 digits
filtered_df['9. ZIP'] = filtered_df['ZIP_str'].str[:5].astype(int)

# Last 4 digits (ZIP+4)
filtered_df['zip+4'] = filtered_df['ZIP_str'].str[5:]

# Replace empty ZIP+4 values with '0' and convert to int
filtered_df['zip+4'] = filtered_df['zip+4'].replace('', '0').astype(int)

# Drop the helper column
filtered_df = filtered_df.drop(columns=['ZIP_str'])

# Get the list of columns
cols = list(filtered_df.columns)

# Remove 'ZIP+4' from current position
cols.remove('zip+4')

# Find the index of '9. ZIP'
zip_index = cols.index('9. ZIP')

# Insert 'ZIP+4' right after '9. ZIP'
cols.insert(zip_index + 1, 'zip+4')

# Reorder the DataFrame
filtered_df = filtered_df[cols]

filtered_df
```
